```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

%matplotlib inline 
```

```python
%ls
```

```python
h = pd.read_csv("cnn2_NET1_history.csv")
```

```python
h.head()
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(2,1,1)
plt.plot(h['loss'], color='blue', marker='o', label='training')
plt.plot(h['val_loss'], color='red', marker='o', label='validation')
plt.legend(fontsize=14)
plt.title("Loss curves",fontsize=14)
plt.ylabel("Binary Crossentropy",fontsize=14)
plt.xlabel("Epoch",fontsize=14)
#plt.show()


plt.figure(figsize=(10, 10))
plt.subplot(2,1,2)
plt.plot(h['accuracy'], color='blue', marker='o', label='training')
plt.plot(h['val_accuracy'], color='red', marker='o', label='validation')
plt.legend(fontsize=14)
plt.title("Accuracy curves",fontsize=14)
plt.ylabel("Accuracy",fontsize=14)
plt.xlabel("Epoch",fontsize=14)
plt.show()

plt.savefig("ann1_acc_loss_curves.png")
```

```python
plt.figure(figsize=(10, 7))
plt.plot(h['auc'], color='blue', marker='o', label='training')
plt.plot(h['val_auc'], color='red', marker='o', label='validation')
plt.legend(fontsize=14)
plt.title("AUC  Curves",fontsize=14)
plt.ylabel("AUC (Precision-Recall)", fontsize=14)
plt.xlabel("Epoch",fontsize=14)
```
