# This notebook visualizes the PR curves of our 3 deep learning models 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb
%matplotlib inline
```

```python
%ls
```

```python
from sklearn.metrics import precision_recall_curve, auc
```

```python
df_cnn1 = pd.read_csv('CNN1_PR_info.csv')
df_cnn2 = pd.read_csv('CNN2_PR_info.csv')
df_ann1 = pd.read_csv('ANN1_PR_info.csv')
```

```python
df_cnn1.head()
```

```python
indcnn1 = np.where(df_cnn1.dist == df_cnn1.dist.min())[0][0]
indcnn2 = np.where(df_cnn2.dist == df_cnn2.dist.min())[0][0]
indann1 = np.where(df_ann1.dist == df_ann1.dist.min())[0][0]


pr_auc_cnn1=auc(df_cnn1['Rec'], df_cnn1['Prec'])
pr_auc_cnn2=auc(df_cnn2.Rec, df_cnn2.Prec)
pr_auc_ann1=auc(df_ann1.Rec, df_ann1.Prec)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(df_cnn1.Rec, df_cnn1.Prec, color="magenta", label="CNN-1 (AUC = {:.3f})".format(df_cnn1.AUC[0]))
plt.plot(df_cnn2.Rec, df_cnn2.Prec, color="crimson", label="CNN-2 (AUC = {:.3f})".format(df_cnn2.AUC[0]))
plt.plot(df_ann1.Rec, df_ann1.Prec, color="darkorange", label="ANN (AUC = {:.3f})".format(df_ann1.AUC[0]))
plt.axhline(0.25, color='k', linestyle='-.', label='Simulated Proportion')
plt.xlabel("Recall", fontsize=14)
plt.ylabel("Precision", fontsize=14)
plt.title("Precision Recall curves", fontsize=14)
plt.legend(fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(df_cnn1.Rec, df_cnn1.Prec, color="magenta", label="CNN-1 (AUC = {:.3f})".format(df_cnn1.AUC[0]))
plt.plot(df_cnn1.Rec[indcnn1], df_cnn1.Prec[indcnn1], color="magenta", 
         label="Threshold = {:.3f}".format(df_cnn1.thres[indcnn1]), marker='o', markersize=10, linestyle='')
plt.axhline(df_cnn1.Prec[indcnn1], color="magenta", linestyle=':', alpha=0.35)
plt.axvline(df_cnn1.Rec[indcnn1], color='magenta', linestyle=':', alpha=0.35)


plt.plot(df_cnn2.Rec, df_cnn2.Prec, color="crimson", label="CNN-2 (AUC = {:.3f})".format(df_cnn2.AUC[0]))
plt.plot(df_cnn2.Rec[indcnn2], df_cnn2.Prec[indcnn2], color="crimson", 
         label="Threshold = {:.3f}".format(df_cnn2.thres[indcnn2]), marker='o', markersize=10, linestyle='')

plt.axhline(df_cnn2.Prec[indcnn2], color="crimson", linestyle=':', alpha=0.35)
plt.axvline(df_cnn2.Rec[indcnn2], color='crimson', linestyle=':', alpha=0.35)



plt.plot(df_ann1.Rec, df_ann1.Prec, color="darkorange", label="ANN (AUC = {:.3f})".format(df_ann1.AUC[0]))
plt.plot(df_ann1.Rec[indann1], df_ann1.Prec[indann1], color="orange", 
         label="Threshold = {:.3f}".format(df_ann1.thres[indann1]), marker='o', markersize=10, linestyle='')


plt.axhline(df_ann1.Prec[indann1], color="darkorange", linestyle=':', alpha=0.35)
plt.axvline(df_ann1.Rec[indann1], color='darkorange', linestyle=':', alpha=0.35)



plt.xlabel("Recall", fontsize=14)
plt.ylabel("Precision", fontsize=14)
plt.title("Precision Recall curves", fontsize=14)
plt.legend(fontsize=13)
plt.xlim(0.8, 1)
plt.ylim(0.8, 1)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(df_cnn1.Rec, df_cnn1.Prec, color="deeppink", label="CNN-1 (AUC = {:.3f})".format(pr_auc_cnn1), linewidth=1.5)
plt.plot(df_cnn1.Rec[indcnn1], df_cnn1.Prec[indcnn1], color="darkcyan", 
         label="Threshold = {:.3f}".format(df_cnn1.thres[indcnn1]), marker='o', markersize=10, linestyle='')
plt.axhline(df_cnn1.Prec[indcnn1], color="darkcyan", linestyle='-.')
plt.axvline(df_cnn1.Rec[indcnn1], color='darkcyan', linestyle='-.')


plt.xlabel("Recall", fontsize=14)
plt.ylabel("Precision", fontsize=14)
plt.title("Precision Recall curves", fontsize=14)
plt.legend(fontsize=13)
plt.xlim(0.68, 1)
plt.ylim(0.68, 1)
plt.show()
```
