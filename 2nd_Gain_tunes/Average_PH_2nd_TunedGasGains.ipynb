```python
import numpy as np 
import matplotlib.pyplot as plt 
from avgph import *
```

```python
import pandas as pd
```

### load in the tuned sim data 

```python
df4000 = pd.read_csv('Clean_10ksimsigs_xegain4000.csv')

df3600 = pd.read_csv('Clean_5ksimsigs_xegain3600.csv')

df3200 = pd.read_csv('Clean_5ksimsigs_xegain3200.csv')
df2800 = pd.read_csv('Clean_5ksimsigs_xegain_2800.csv')
df2400 = pd.read_csv('Clean_5ksimsigs_xegain2400.csv')
df2000 = pd.read_csv('Clean_5ksimsigs_xegain2000.csv')
df1600 = pd.read_csv('Clean_5ksimsigs_xegain1600.csv')
df1200 = pd.read_csv('Clean_5ksimsigs_xegain1200.csv')

df800 = pd.read_csv('Clean_5ksimsigs_xegain800.csv')
df400 = pd.read_csv('Clean_5ksimsigs_xegain400.csv')

```

## load the real data 

```python
real = pd.read_csv('../Stephan_Raw_Data/Clean_realsigs_LHC22o.csv')
```

```python
real.shape
```

```python
tr, adcr = pharray(real)
```

```python
t4000, adc4000 = pharray(df4000)

t3600, adc3600 = pharray(df3600)

t3200, adc3200 = pharray(df3200)
t2800, adc2800 = pharray(df2800)
t2400, adc2400 = pharray(df2400)
t2000, adc2000 = pharray(df2000)
t1600, adc1600 = pharray(df1600)
t1200, adc1200 = pharray(df1200)


t800, adc800 = pharray(df800)
t400, adc400 = pharray(df400)


```

```python
plt.figure(figsize=(15, 7))
plt.plot(t4000, adc4000, marker='.', label='gain=4000', linestyle=':')
plt.plot(t3600, adc3600, marker='.', label='gain=3600', linestyle=':')
plt.plot(t3200, adc3200, marker='.', label='gain=3200', linestyle=':')
plt.plot(t2800, adc2800, marker='.', label='gain=2800', linestyle=':')
plt.plot(t2400, adc2400, marker='.', label='gain=2400', linestyle=':')
plt.plot(t2000, adc2000, marker='.', label='gain=2000', linestyle=':')
plt.plot(t1600, adc1600, marker='.', label='gain=1600', linestyle=':')
plt.plot(t1200, adc1200, marker='.', label='gain=1200', linestyle=':')
plt.plot(t800, adc800, marker='.', label='gain=800', linestyle=':')
plt.plot(t400, adc400, marker='.', label='gain=400', linestyle=':')

plt.plot(tr, adcr, marker='o', label='Run 526689', color='blue')


plt.legend()
plt.title("Average Pulse Height", fontsize=16)
plt.xlabel("Time bin (100ns)", fontsize=16)
plt.ylabel("ADC", fontsize=16)
plt.show()
```

## Lets now plot the ratios of each tuned gain PH curve to the real one 

```python
r4000 = np.array(adcr)/np.array(adc4000)

r3600 = np.array(adcr)/np.array(adc3600)

r3200 = np.array(adcr)/np.array(adc3200)
r2800 = np.array(adcr)/np.array(adc2800)
r2400 = np.array(adcr)/np.array(adc2400)
r2000 = np.array(adcr)/np.array(adc2000)
r1600 = np.array(adcr)/np.array(adc1600)
r1200 = np.array(adcr)/np.array(adc1200)

r800 = np.array(adcr)/np.array(adc800)
r400 = np.array(adcr)/np.array(adc400)
```

```python
plt.figure(figsize=(15, 7))
plt.axhline(y=1, color='black', linestyle='--', label="Ideal Ratio")
plt.plot(t4000, r4000, marker='.', label='gain=4000', linestyle=':')

plt.plot(t3600, r3600, marker='.', label='gain=3600', linestyle=':')

plt.plot(t3200, r3200, marker='.', label='gain=3200', linestyle=':')
plt.plot(t2800, r2800, marker='.', label='gain=2800', linestyle=':')
plt.plot(t2400, r2400, marker='.', label='gain=2400', linestyle=':')
plt.plot(t2000, r2000, marker='.', label='gain=2000', linestyle=':')
plt.plot(t1600, r1600, marker='.', label='gain=1600', linestyle=':')
plt.plot(t1200, r1200, marker='.', label='gain=1200', linestyle=':')

plt.plot(t800, r800, marker='.', label='gain=800', linestyle=':')
plt.plot(t400, r400, marker='.', label='gain=400', linestyle=':')

plt.legend()
plt.title("Average Pulse Height Ratios", fontsize=16)
plt.ylabel("$\\frac{Real}{Sim}$", fontsize=16)
plt.xlabel("Time bin (100ns)", fontsize=16)
plt.show()
```
